```python
# DecodeLabs AI Internship — Project 1: Rule-Based AI Chatbot
# Architecture: Input -> Sanitize -> Process (dict lookup) -> Output

from datetime import datetime

# Dictionary lookup instead of if-elif ladder: O(1) lookup vs O(n) chain,
# and it scales cleanly if more intents get added later.
KNOWLEDGE_BASE = {
    "hello": "Hi there! How can I help you today?",
    "hi": "Hello! What's on your mind?",
    "hey": "Hey! Good to see you.",
    "how are you": "I'm just a bunch of if-else logic, but I'm doing great!",
    "what is your name": "I'm DecodeBot, your friendly rule-based assistant.",
    "what can you do": "Right now I can chat using a fixed set of rules. No learning yet, just logic.",
    "help": "Try greeting me, asking my name, or just say 'bye' when you're done.",
    "thanks": "You're welcome!",
    "thank you": "Anytime!",
}

# Words that end the conversation (the "kill command" from the spec)
EXIT_COMMANDS = {"exit", "quit", "bye", "goodbye"}

FALLBACK_RESPONSE = "I do not understand. Try saying 'help' to see what I know."


def sanitize_input(raw_text):
    """Normalize whatever the user typed so dictionary matching is reliable."""
    return raw_text.lower().strip()


def get_response(clean_text):
    """Single atomic lookup-plus-fallback operation."""
    return KNOWLEDGE_BASE.get(clean_text, FALLBACK_RESPONSE)
```

```python
# Run this cell live and actually type to the bot.
# Note: this will only work in an interactive session (editing the notebook
# yourself). If you use Kaggle's "Save & Run All (Commit)" option it runs
# headlessly with no stdin, so input() will fail — that's expected.
# To preserve your output: chat with it here, then use
# "Save Version -> Quick Save" (NOT Save & Run All) so this output sticks.

from datetime import datetime

transcript = []
print("DecodeBot: Hello! I'm a rule-based chatbot. Type 'bye' to exit.\n")

while True:
    try:
        raw_text = input("You: ")
    except EOFError:
        print("DecodeBot: (no input stream available — run this cell interactively)")
        break

    clean_text = sanitize_input(raw_text)
    transcript.append(f"You: {raw_text}")

    if clean_text in EXIT_COMMANDS:
        print("DecodeBot: Goodbye! Talk soon.")
        transcript.append("DecodeBot: Goodbye! Talk soon.")
        break

    reply = get_response(clean_text)
    print(f"DecodeBot: {reply}")
    transcript.append(f"DecodeBot: {reply}")

# Save the real conversation you just had
output_path = "/kaggle/working/chatbot_conversation_log.txt"
with open(output_path, "w") as f:
    f.write(f"DecodeBot Conversation Log — {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write("=" * 50 + "\n")
    f.write("\n".join(transcript))

print(f"\nTranscript saved to {output_path}")
```
